```python
def toxicityColors(tweet, word_to_replace):
  colors1 = ["negr", "blanc", "roj", "amarill", "morad"]
  var1 = ["a", "as", "o", "os"]
  colors2 = ["verde", "naranja", "azul"]
  plurals  = ["s", "es"]

  colors_v = {}
  for i in colors1:
    for j in var1:
      v = pipe((tweet.replace(word_to_replace, i) + j), return_all_scores=True)[0]
      colors_v.update({i+j: (v[0]['score'], v[1]['score'])})
    for i in colors2:
      v = pipe((tweet.replace(word_to_replace, i)), return_all_scores=True)[0]
      colors_v.update({i: (v[0]['score'], v[1]['score'])})
      if i != "azul":
        v = pipe((tweet.replace(word_to_replace, i+plurals[0])), return_all_scores=True)[0]
        colors_v.update({i+plurals[0]: (v[0]['score'], v[1]['score'])})
      else:
        v = pipe((tweet.replace(word_to_replace, i+plurals[1])), return_all_scores=True)[0]
        colors_v.update({i+plurals[1]: (v[0]['score'], v[1]['score'])})     
  return colors_v
```
