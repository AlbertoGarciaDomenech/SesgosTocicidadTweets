```python
lideres = ["Casado", "Ayuso", "Abascal", "Monasterio" ,"Sanchez", "Sánchez", "Iglesias", "Yolanda", "Arrimadas", "Rivera"]
ids = ["@pablocasado", "@IdiazAyuso", "@Santi_ABASCAL", "@monasterioR", "@sanchezcastejon", "@Yolanda_Diaz", "@InesArrimadas"] #Pablo iglesias y Albert Rivera salen demasiado poco sus @
nombresLideres  = set(lideres).union(set(ids))
nombresLideres
```

```python
def variacionToxicidad(df, politician, var=0.10, very_tox=0, key_words=nombresLideres):
  lead = {i : 0 for i in lideres}
  tweetsNumber = 0

  for idx, row in df.iterrows():
    if (row.original_name == politician):
      tweetsNumber += 1
      if very_tox == 0:
        if row.most_toxic_leader_toxicity > (row.toxic_original + var):
          ntox = lead.get(row.most_toxic_leader)
          ntox += 1
          lead.update({row.most_toxic_leader: ntox})
      else:
        if row.most_verytoxic_leader_toxicity > (row.very_toxic_original + var):
          ntox = lead.get(row.most_verytoxic_leader)
          ntox += 1
          lead.update({row.most_verytoxic_leader: ntox})

  print('Tweets: ', tweetsNumber)
  for p in lead:
    print('\t -', p, ': ', lead.get(p))
```

```python
tweetsLeaders = df.copy()
# lideres = ["Casado", "Ayuso", "Abascal", "Monasterio" ,"Sanchez", "Sánchez", "Iglesias", "Yolanda", "Arrimadas", "Rivera"]
```

```python
for i in lideres:
  print(i)
  variacionToxicidad(tweetsLeaders,politician = i)
  variacionToxicidad(tweetsLeaders,politician = i, very_tox=1)
```
