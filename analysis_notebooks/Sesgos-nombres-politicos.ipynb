```python
from re import A
def variacionToxicidad(df, politician, var=0.10, very_tox=0):
  lead = {i : 0 for i in nombres}
  tweetsNumber = 0
  tweetsNumberV = 0

  for idx, row in df.iterrows():
    if (row.original_name == politician):
      tweetsNumber += 1
      if very_tox == 0:
        if row.most_toxic_leader_toxicity > (row.toxic_original + var):
          ntox = lead.get(row.most_toxic_leader)
          ntox += 1
          lead.update({row.most_toxic_leader: ntox})
      else:
        if row.most_verytoxic_leader_toxicity > (row.very_toxic_original + var):
          ntox = lead.get(row.most_verytoxic_leader)
          ntox += 1
          lead.update({row.most_verytoxic_leader: ntox})

  if very_tox == 0:
    print('Numero de tweets con ', politician,'como nombre original: ', tweetsNumber)
    print("Numero de veces que sustituyendo a ", politician, " por cada politico resulta mas toxico en la escala toxic: ")
  else:
    print("--------------------------------------------------------------------------------------------------------------")
    print("Numero de veces que sustituyendo a ", politician, " por cada politico resulta mas toxico en la escala very toxic: ")

  for p in lead:
    if(p != politician):
      print('\t -', p, ': ', lead.get(p))  
print("--------------------------------------------------------------------------------------------------------------")
```

```python
tweetsLeaders = df.copy()
# lideres = ["Casado", "Ayuso", "Abascal", "Monasterio" ,"Sanchez", "Sánchez", "Iglesias", "Yolanda", "Arrimadas", "Rivera"]
```

```python
for i in nombres:
  variacionToxicidad(tweetsLeaders,politician = i)
  variacionToxicidad(tweetsLeaders,politician = i, very_tox=1)
```
